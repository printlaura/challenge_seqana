```python
df = pd.get_dummies(df, columns=['year'])
df
```

### Linear regression

```python
X = df.loc[:, df.columns != 'soc_stock_t_ha'] #features
y = df['soc_stock_t_ha'] #target
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=324)
```

```python
scale = preprocessing.StandardScaler()
scale.fit(X_train)
X_train = scale.transform(X_train)
```

```python
print(X_train.max())
print(X_train.min())
```

```python
lin_regressor = LinearRegression()

lin_regressor.fit(X_train, y_train)
```

```python
X_test = scale.transform(X_test)
y_prediction = lin_regressor.predict(X_test)
y_result = y_prediction - y_test

y_result
```

```python
RMSE = sqrt(mean_squared_error(y_true = y_test, y_pred = y_prediction))
RMSE
```

```python
lin_regressor.score(X_test, y_test)
```
